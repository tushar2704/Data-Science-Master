**Case:- 3) Build a DT model for social adv data by including "Gender" as an input variable (x var) along with the other two x vars.**

```python
#Loading the required data

social_data = pd.read_csv('/kaggle/input/social-network/Social_Network_Ads.csv')
```

```python
#Checking the loaded data
social_data.head()
#pd.set_option('display.max_columns', None)
```

**Basics Checks**

```python
#Checking for all columns names
social_data.columns
```

```python
#Checking for shape of dataset
social_data.shape, print(social_data.info())
```

```python
#Cheking statistics
social_data.describe(include="all") 
```

```python
#convert categorical data('Gender') to binary data('0 & 1')
social_data_1 = pd.get_dummies(social_data["Gender"])
social_data_1
```

```python
#Using pd.get_dummies
social_data_converted = pd.concat((social_data, social_data_1), axis=1)
social_data_converted = social_data_converted.drop(['Gender'], axis=1)
social_data_converted = social_data_converted.drop(['Female'], axis=1)

social_data_converted
```

```python
#Seprating for features and target variable
X = social_data_converted[['Age','EstimatedSalary','Male']].values
y = social_data_converted[['Purchased']].values
```

```python
#Checking for columns
#print(X), print(y)
```

```python
#Splitting train and test sets

X_train, X_test, y_train, y_test  = train_test_split(X,y, test_size=0.30,random_state=123)
```

```python
#Checking X_train, X_test, y_train, y_test shapes
print(X_train.shape), print(X_test.shape), print(y_train.shape), print(y_test.shape)
```

**Feature Scaling**

```python
sc = StandardScaler() # creating the instance
```

```python
#Standardizing the features
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
#Checking X_train, X_test
print(X_train), print(X_test)
```

```python
#Importing the model DecisionTreeClassifier as target will be binary

from sklearn.tree import DecisionTreeClassifier
```

**Building the DT Model using the Training data**

**Here we create a DT classifier object.**

```python
#Creting the insatance
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 123)
```

```python
#Fitting the model on trainig data
classifier.fit(X_train, y_train)
```

```python
#Predicting on "Test Data"
y_pred = classifier.predict(X_test)
print(y_pred)
```

```python
#Checking for 
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
#Getting the confusion matrix and accuracy score
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

**Classification_report for Build a DT model for social adv data by including "Gender" as an input variable (x var) along with the other two x vars**

```python
#printing Classification_report
print(classification_report(y_test, y_pred))
```

**Building the DT Model using the Training data
