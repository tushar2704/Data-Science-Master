<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:17px;border-radius:20px">Importing the required Libraries
    
   </p>

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, classification_report
```

<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:17px;border-radius:20px">
    Loading the data
   </p>

```python
#Reading the data file
social_data = pd.read_csv('/kaggle/input/social-network/Social_Network_Ads.csv')
```

```python
#Printing the head for data
social_data.head(11)
```

```python
#Checking info for the data
social_data.info()
```

<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:17px;border-radius:20px">
    Selecting data using iloc - Here we select all rows but only required columns
   </p>

```python
#Seprating for features and target variable
X = social_data.iloc[:, [2,3]].values
y = social_data.iloc[:,4].values
```

```python
#Checking for columns
#print(X), print(y)
```

<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:17px;border-radius:20px">
    Splitting the dataset into the Training set and Test set
   </p>

```python
#Splitting train and test sets
X_train, X_test, y_train, y_test  = train_test_split(X,y, test_size=0.3,random_state=0)
```

```python
#Checking X_train, X_test, y_train, y_test shapes
print(X_train.shape), print(X_test.shape), print(y_train.shape), print(y_test.shape)
```

<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:17px;border-radius:20px">
    Feature Scaling
    <br><br>Here we only scale X_train and X_test
   </p>

```python
sc = StandardScaler() # creating the instance
```

```python
#Standardizing the features
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
#Checking X_train, X_test
print(X_train), print(X_test)
```

<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:17px;border-radius:20px">
    Building the DT Model using the Training data
    <br><br>Here we create a DT classifier object

```python
#Fitting the model on trainig data
#classifier.fit(X_train, y_train)
```

```python
#Creting the insatance
#classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
```

<p style="width: 700px;padding: 20px;background: papayawhip;border-radius:10px;font-size:17px;border-radius:20px">
    Building the model - Random Forest Classifier
   </p>

```python
random_cal = RandomForestClassifier(n_estimators=25, criterion='entropy', random_state=123)
```

```python
#Fitting the model on trainig data
random_cal.fit(X_train, y_train)
```

```python
#Predicting on "Test Data"
y_pred = random_cal.predict(X_test)
print(y_pred)
```

```python
#Checking for 
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
#Getting the confusion matrix and accuracy score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
#Getting the confusion matrix and accuracy score
cm=confusion_matrix(y_test, y_pred)

print(cm)
accuracy_score(y_test, y_pred)
```

```python
#printing Classification_report
print(classification_report(y_test, y_pred))
```
